```python
import pandas as pd

# Load CSV file
csv_path = "../../Datasets/dataset.csv"
df = pd.read_csv(csv_path)

df['has_error'] = df['has_error'].astype(bool)

# Group the data by 'dataset_name' and compute statistics
analysis = df.groupby('dataset_name').agg(
    total_files=('file_name', 'count'),  # Total files
    error_files=('has_error', 'sum'),    # Total files with error
    error_rate=('has_error', 'mean')     # Error rate
).reset_index()

# Convert error rate to percentage
analysis['error_rate'] = analysis['error_rate'] * 100

# Display the analysis
print("Statistical Analysis by Dataset:")
print(analysis)

# Save the analysis to a CSV file (optional)
output_path = "dataset_analysis.csv"
analysis.to_csv(output_path, index=False)
print(f"Analysis saved to: {output_path}")
```
